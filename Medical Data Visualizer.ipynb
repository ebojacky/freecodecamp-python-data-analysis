```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/Users/ebojackson/Library/CloudStorage/OneDrive-Personal/Documents/My Python/FreeCodeCamp/data-analysis-with-python/projects/medical_examination.csv")
```

```python
df["overweight"] = df["weight"]/((df["height"]/100)**2)

#To determine if a person is overweight, first calculate their BMI by 
#dividing their weight in kilograms by the square of their height in meters. 
#If that value is > 25 then the person is overweight. 
#Use the value 0 for NOT overweight and the value 1 for overweight.

df.loc[df["overweight"] <= 25,"overweight"]=int(0)
df.loc[df["overweight"] > 25,"overweight"]=int(1)
```

```python
df["overweight"] = df["overweight"].astype("int")
df
```

```python
df.loc[df["gluc"] <2,"gluc"]=0
```

```python
df.loc[df["gluc"] >=2,"gluc"]=1
```

```python
df.loc[df["cholesterol"] <2,"cholesterol"]=0
```

```python
df.loc[df["cholesterol"] >=2,"cholesterol"]=1
```

```python
#option 1

df_cat = pd.melt(df, id_vars=["cardio"], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
print(df_cat)
#df_cat["total"] = 1
df_cat = df_cat.groupby(["cardio","variable","value"],as_index=False).value_counts()
print(df_cat)
sns.catplot(data=df_cat, x="variable", y="count", hue="value", kind="bar", col="cardio")
```

```python
#option 2

df_cat = pd.melt(df, id_vars=["cardio"], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
print(df_cat)
df_cat = pd.DataFrame(df_cat.groupby(['cardio', 'variable', 'value'])["value"].count()).rename(columns={'value': 'total'}).reset_index()

print(df_cat)
print(type(df_cat))
sns.catplot(data=df_cat, x="variable", y="total", hue="value", kind="bar", col="cardio")
```
